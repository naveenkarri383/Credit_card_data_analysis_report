```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('mysql+pymysql://root:root@127.0.0.1:3306/credit_card_details')
```

```python
data_credit_card_info.to_sql('data_credit_card_info', con=engine, if_exists='replace', index=False)
```

```python
query = "SELECT * FROM data_credit_card_info"
data_credit_card_info = pd.read_sql(query, con=engine)
```

```python
## Connecting to customers details database
engine = create_engine('mysql+pymysql://root:root@127.0.0.1:3306/credit_card_details')
```

```python
# loading data to database
data_customers_info.to_sql('customer_details', con=engine, if_exists='replace', index=False)
```

```python
query = "SELECT * FROM customer_details"
data_customers_info = pd.read_sql(query, con=engine)
```
